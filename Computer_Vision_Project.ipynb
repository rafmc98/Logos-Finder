<h2>Training of YOLOV3</h2>

```python
from kerasyolo3.yolo3.model import preprocess_true_boxes, yolo_body, tiny_yolo_body,yolo_loss
from kerasyolo3.yolo3.utils import get_random_data
import keras.backend as K
import numpy as np
from keras.layers import Input, Lambda
from keras.models import Model
from keras.optimizers import Adam
from keras.callbacks import TensorBoard, ModelCheckpoint, ReduceLROnPlateau,EarlyStopping
#Function used to create the architecture of the YOLOV3 model
def create_model(input_shape, anchors, num_classes, load_pretrained=False, freeze_body=2,
            weights_path='Models/yolo_logoInTheWild_100epochs_stage2.h5'):
    '''create the training model'''
    K.clear_session() # get a new session
    image_input = Input(shape=(None,None,3))
    h, w = input_shape
    num_anchors = len(anchors)
    y_true = [Input(shape=(h//{0:32, 1:16, 2:8}[l], w//{0:32, 1:16, 2:8}[l], \
        num_anchors//3, num_classes+5)) for l in range(3)]
    model_body = yolo_body(image_input, num_anchors//3, num_classes)
    print('Create YOLOv3 model with {} anchors and {} classes.'.format(num_anchors, num_classes))
    if load_pretrained:
        model_body.load_weights(weights_path, by_name=True, skip_mismatch=True)
        print('Load weights {}.'.format(weights_path))
        if freeze_body in [1, 2]:
            num = (185, len(model_body.layers)-3)[freeze_body-1]
            for i in range(num): model_body.layers[i].trainable = False
            print('Freeze the first {} layers of total {} layers.'.format(num, len(model_body.layers)))
    model_loss = Lambda(yolo_loss, output_shape=(1,), name='yolo_loss',
        arguments={'anchors': anchors, 'num_classes': num_classes, 'ignore_thresh': 0.5})(
        [*model_body.output, *y_true])
    model = Model([model_body.input, *y_true], model_loss)
    return model
```

```python
import numpy as np
#Function used to get the array of the classes for the YOLO training (in this case only LOGO class) from the class file
def get_classes(classes_path='Dataset/LogosInTheWild/data_classes.txt'):
    '''loads the classes'''
    with open(classes_path) as f:
        class_names = f.readlines()
    class_names = [c.strip() for c in class_names]
    return class_names
#Function used to get the anchors for the YOLO training from the anchor file
def get_anchors(anchors_path='kerasyolo3/model_data/yolo_anchors.txt'):
    '''loads the anchors from a file'''
    with open(anchors_path) as f:
        anchors = f.readline()
    anchors = [float(x) for x in anchors.split(',')]
    return np.array(anchors).reshape(-1, 2)
```

```python
#Callbacks for the training
checkpoint = ModelCheckpoint('ep{epoch:03d}-loss{loss:.3f}-val_loss{val_loss:.3f}.h5',monitor='val_loss', save_weights_only=True, save_best_only=True, period=3)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, verbose=1)
early_stopping = EarlyStopping(monitor='val_loss', min_delta=0, patience=10, verbose=1)
```

```python
#Data generator for the images and the related ground truth boxes in the dataset
def data_generator(annotation_lines, batch_size, input_shape, anchors, num_classes):
    '''data generator for fit_generator'''
    n = len(annotation_lines)
    i = 0
    while True:
        image_data = []
        box_data = []
        for b in range(batch_size):
            if i==0:
                np.random.shuffle(annotation_lines)
            image, box = get_random_data(annotation_lines[i].replace("\\","/"), input_shape, random=True)
            image_data.append(image)
            box_data.append(box)
            i = (i+1) % n
        image_data = np.array(image_data)
        box_data = np.array(box_data)
        y_true = preprocess_true_boxes(box_data, input_shape, anchors, num_classes)
        yield [image_data, *y_true], np.zeros(batch_size)
#Wrapper for the data generator. Used to divide trainig set from validation set
def data_generator_wrapper(annotation_lines, batch_size, input_shape, anchors, num_classes):
    n = len(annotation_lines)
    if n==0 or batch_size<=0: return None
    return data_generator(annotation_lines, batch_size, input_shape, anchors, num_classes)
```

```python
#function used to get the two arrays containing images and ground truth box for the evaluation
def get_evaluation_list(evalation_dataset_path="Dataset/LogoDet3k/voc_format/"):
    image_set=set([])
    image_list=[]
    box_list=[]
    n_samples=0
    n_classes=len(os.listdir(evalation_dataset_path))
    while len(image_set)<4000:
        found=False
        while not found:
            random_class=random.randint(0,n_classes-1)
            class_path=os.path.join(evalation_dataset_path,os.listdir(evalation_dataset_path)[random_class])
            n_samples+=1
            class_images=len(os.listdir(class_path))
            if class_images==0:
                continue
            random_image=random.randint(0,class_images-1)
            file_path=os.path.join(class_path,os.listdir(class_path)[random_image])
            if (file_path.split(".")[-1]=='xml'):
                #Xml case
                if os.path.isfile(file_path.split(".")[0]+".jpg"):
                    img_path=file_path.split(".")[0]+".jpg"
                    try:
                        boxes=get_boxes(file_path)
                    except:
                        continue
                    if len(boxes)==1 and img_path not in image_set:
                        image_set.add(img_path)
                        image_list.append(img_path)
                        box_list.append(boxes)
                        found=True
                elif os.path.isfile(file_path.split(".")[0]+".png"):
                    img_path=file_path.split(".")[0]+".png"
                    try:
                        boxes=get_boxes(file_path)
                    except:
                        continue
                    if len(boxes)==1 and img_path not in image_set:
                        image_set.add(img_path)
                        image_list.append(img_path)
                        box_list.append(boxes)
                        found=True
            else:
                if os.path.isfile(file_path.split(".")[0]+".xml"):
                    image_path=file_path
                    try:
                        boxes=get_boxes(file_path)
                    except:
                        continue
                    if len(boxes)==1 and img_path not in image_set:
                        image_set.add(img_path)
                        image_list.append(img_path)
                        box_list.append(boxes)
                        found=True
    return image_list,box_list
#function used to retrieve the box points from the xml file
def get_boxes(filepath):
    with open(filepath, "r") as file:
        # Read each line in the file, readlines() returns a list of lines
        content = file.readlines()
        # Combine the lines in the list into a string
        content = str("".join(content))
        bs_content = bs(content, "lxml")
        result = bs_content.find_all("bndbox")
        l=[]
        for box in result:
            b=[]
            b.append(int(bs_content.find("xmin").text))
            b.append(int(bs_content.find("ymin").text))
            b.append(int(bs_content.find("xmax").text))
            b.append(int(bs_content.find("ymax").text))
            l.append(b)
        return l    
```

```python
#function used to make the predictions using the yolo model
def detect_logos(yolo,img_path):
    img=Image.open(img_path)
    if img.mode != "RGB":
        img = image.convert("RGB")
    image_array = np.array(img)
    predictions=yolo.detect_image(img)
    return predictions
```

```python
#function for the IOU calcolous given the ground truth box and the prediction box
def bb_intersection_over_union(boxA, boxB):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    # compute the area of intersection rectangle
    interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)
    # compute the area of both the prediction and ground-truth
    # rectangles
    boxAArea = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
    boxBArea = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)
    # compute the intersection over union by taking the intersection
    # area and dividing it by the sum of prediction + ground-truth
    # areas - the interesection area
    iou = interArea / float(boxAArea + boxBArea - interArea)
    # return the intersection over union value
    return iou
```

```python
import os
import tensorflow as tf
import keras
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, Flatten,\
                         Conv2D, MaxPooling2D
from keras.layers import BatchNormalization
from keras import regularizers
from keras import optimizers
from PIL import Image
from keras import applications
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, Flatten,\
                         Conv2D, MaxPooling2D, AveragePooling2D
from keras.layers import BatchNormalization
from keras import regularizers
from keras.optimizers import Adam
from keras import applications
import numpy as np
import PIL
from sklearn.utils import class_weight
from keras.applications.inception_v3 import InceptionV3
from keras import layers
#Loading of the dataset in memory and preprocessing
def loadDataset(modelName=None):
    datadir = 'Dataset/LogosInTheWild/brandROIs_balanced'
    batch_size = 32
    if modelName==None:
        print("Using generic preprocessing")
        train_datagen = ImageDataGenerator(
        rescale = 1. / 255,\
        zoom_range=0.1,\
        rotation_range=10,\
        width_shift_range=0.1,\
        height_shift_range=0.1,\
        horizontal_flip=True,\
        vertical_flip=False,
        validation_split=0.2)
    else:
        if modelName=="inception":
            preprocessing=applications.inception_v3.preprocess_input
            print("Using InceptionV3 preprocessing")      
        elif modelName=='resnet':
            preprocessing=applications.resnet_v2.preprocess_input
            print("Using ResNet50V2 preprocessing")
        elif modelName=='vgg16':
            preprocessing=applications.vgg16.preprocess_input
            print("Using Vgg16 preprocessing")        
        train_datagen = ImageDataGenerator(
        preprocessing_function=preprocessing,
        horizontal_flip=True,\
        vertical_flip=False,
        validation_split=0.2)
    train_generator = train_datagen.flow_from_directory(
        directory=datadir,
        target_size=(224, 224),
        color_mode="rgb",
        batch_size=batch_size,
        class_mode="categorical",
        subset='training',
        seed=1)
    test_generator = train_datagen.flow_from_directory(
        directory=datadir,
        target_size=(224, 224),
        color_mode="rgb",
        batch_size=batch_size,
        class_mode="categorical",
        subset="validation",
        seed=1,
        shuffle=False)
    num_samples = train_generator.n
    num_classes = train_generator.num_classes
    input_shape = train_generator.image_shape
    classnames = [k for k,v in train_generator.class_indices.items()]
    return train_generator,test_generator
```

```python
#Retrivial of the model architecture. You must choose between InceptionV3,Resnet50V2,VGG16
def get_model(name,n_classes,input_shape):
    if name=='inception':
        model = applications.inception_v3.InceptionV3(
                                     include_top = True,
                                     weights = None,
                                     input_shape =input_shape,
                                     classes=n_classes)
    elif name=='resnet':
        model = applications.resnet_v2.ResNet50V2(
                                     include_top = True,
                                     weights = None,
                                     input_shape =input_shape,
                                     classes=n_classes)
    elif name=='vgg16':
        model = applications.vgg16.VGG16(
                                     include_top = True,
                                     weights = None,
                                     input_shape =input_shape,
                                     classes=n_classes)
    for layer in model.layers:
        layer.trainable=True
    return model
```

```python
from PIL import Image
#function used to train the model in input and to save the weights
def train_model(model,train_gen,test_gen,ep,filename):
    class_weights = class_weight.compute_class_weight('balanced',
                                                  classes = np.unique(train_gen.classes), 
                                                  y = train_gen.classes)
    class_weights_dict =dict(enumerate(class_weights))
    model.compile(loss='categorical_crossentropy', optimizer=Adam(lr = 0.0001), metrics=['accuracy'])
    steps_per_epoch=train_gen.n//train_gen.batch_size
    val_steps=test_gen.n//test_gen.batch_size+1
    history=model.fit_generator(train_gen, epochs=ep, verbose=1,
                            steps_per_epoch=steps_per_epoch,
                            validation_data=test_gen,
                            validation_steps=val_steps,
                            class_weight=class_weights_dict)
    model.save("Models/"+filename+".h5")
    return history
```

```python
import keras.backend as K
#Metrics callback for evaluate single models
def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```python
#Creation of the architecture for transfer learning
def get_transfer_learning_model(input_shape,n_classes,basemodel='inception'):

    if basemodel=='inception':
        featurizer = applications.inception_v3.InceptionV3(
                                     include_top = False,
                                     weights = 'imagenet',
                                     input_shape =input_shape)
    elif basemodel=='resnet':
        featurizer = applications.resnet_v2.ResNet50V2(
                                     include_top = False,
                                     weights = 'imagenet',
                                     input_shape =input_shape)
    elif basemodel=='vgg16':
        featurizer = applications.vgg16.VGG16(
                                     include_top = False,
                                     weights = 'imagenet',
                                     input_shape =input_shape)
    final_model=Sequential([])
    final_model.add(featurizer)
    final_model.add(Flatten())
    final_model.add(layers.Dense(units=256,activation='relu'))
    final_model.add(layers.Dense(units=256,activation='relu'))
    final_model.add(layers.Dense(units=n_classes,activation='softmax'))
    return final_model
    
```

```python
from keras.callbacks import ReduceLROnPlateau
from keras.callbacks import EarlyStopping
#Creation of the callbacks used during training
def prepare_callbacks():
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, verbose=1)
    early_stopping = EarlyStopping(monitor='val_loss', min_delta=0, patience=10, verbose=1)
    return [reduce_lr,early_stopping]

```

```python
#Function used for the training of the model
def train_transfer_model(model,train_gen,test_gen,epoch_first,epoch_last,filename):
    class_weights = class_weight.compute_class_weight('balanced',
                                              classes = np.unique(train_gen.classes), 
                                              y = train_gen.classes)
    class_weights_dict =dict(enumerate(class_weights))
    model.compile(loss='categorical_crossentropy', optimizer=Adam(lr =1e-4), metrics=['accuracy'])
    steps_per_epoch=train_gen.n//train_gen.batch_size
    val_steps=test_gen.n//test_gen.batch_size+1
    cb=prepare_callbacks()
    model.fit_generator(train_gen, epochs=epoch_first, verbose=1,
                            steps_per_epoch=steps_per_epoch,
                            validation_data=test_gen,
                            validation_steps=val_steps,
                            initial_epoch=0,
                            class_weight=class_weights_dict)
    model.layers[0].trainable=True
    print("Starting Fine Tuning step. Featurizer trainable state: "+str(model.layers[0].trainable))
    model.compile(loss='categorical_crossentropy', optimizer=Adam(lr = 1e-6), metrics=['accuracy'])
    model.fit_generator(train_gen, epochs=epoch_last, verbose=1,
                        steps_per_epoch=steps_per_epoch,
                        validation_data=test_gen,
                        validation_steps=val_steps,
                        initial_epoch=epoch_first,
                        class_weight=class_weights_dict,
                        callbacks=cb)
 
    model.save("Models/"+filename+".h5")
    return model
    
```
